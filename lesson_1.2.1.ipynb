#### Занятие 1.2.1 | Графики функций c numpy и matplotlib

Содержание: 
- Графики в matplotlib (np.array)
- Равномерное разбиение отрезка значений (np.linspace)
- Векторные операции
- Режимы работы \%matplotlib tk и \%matplotlib inline

```python
# Импортируем модули
# numpy - удобная и быстрая работа с массивами
# matplotlib - библиотека для построения графиков
import numpy as np
import matplotlib.pylab as plt
```

```python
# Вывод версий библиотек
np.__version__, plt.__version__
```

```python
# Вывод графика
y_array = [1,2,6,2,1]
plt.plot(y_array)
```

```python
# Вывод графика
x_array = np.array([0,1,2,3,4])
y_array = np.array([1,2,6,2,1])
plt.plot(x_array, y_array)
```

```python
# График линейной функции
X = np.linspace(-10,10,1000)
Y = 10*X
plt.plot(X,Y)
```

```python
# График степенной функции
X = np.linspace(-10,10,1000)
Y = X**3
plt.plot(X,Y)
```

```python
# Влияние частоты разбиения отрезка на визуализацию
X_1 = np.linspace(-10,10,3)
X_2 = np.linspace(-10,10,10)
X_3 = np.linspace(-10,10,100)
X_4 = np.linspace(-10,10,1000)
plt.plot(X_1,X_1**3)
plt.plot(X_2,X_2**3)
plt.plot(X_3,X_3**3)
plt.plot(X_4,X_4**3)
```

```python
# График тригонометрической функции
X = np.linspace(-10,10,1000)
plt.plot(X, np.sin(X))
```

```python
X = np.linspace(-10,10,10000)
plt.plot(X, np.sin(10./X))
```

```python
# Переключение в режим работы с отображением графиков в отдельных окнах
%matplotlib tk
```

```python
X = np.linspace(-10,10,10000)
plt.plot(X, np.sin(10./X))
```

```python
# Переключение в режим работы с отображением графиков в jupyter notebook
%matplotlib inline
```

```python
X = np.linspace(-10,10,10000)
plt.plot(X, X**3)
```

```python
# Вывод нескольких графиков в одной ячейке
X = np.linspace(-10,10,10000)
plt.plot(X, X**3)
plt.show()
plt.plot(X, np.sin(X))
plt.show()
plt.plot(X, np.sin(10./X))
plt.show()
```
