#### Занятие 1.2.2 | Графики функций. Цвет, тип и толщина линии

```python
# Импортируем модули
# numpy - удобная и быстрая работа с массивами
# matplotlib - библиотека для построения графиков
import numpy as np
import matplotlib.pylab as plt
```

```python
# Вывод версий библиотек
np.__version__, plt.__version__
```

```python
# График функции y = x^3 и y = x^3 c наложенным шумом
X = np.linspace(-10,10,100)
Y = X**3

Y_noise = Y + np.random.normal(0,100,100)
```

```python
# Стиль графиков по умолчанию
plt.plot(X,Y)
plt.plot(X,Y_noise)
```

```python
# Короткая форма записи для настроек оформления. Цвет линии
plt.plot(X,Y, 'g')
plt.plot(X,Y_noise, 'r')
```

```python
# Короткая форма записи для настроек оформления. Цвет и тип линии
plt.plot(X,Y, 'k-')
plt.plot(X,Y_noise, 'r--')
```

```python
# Короткая форма записи для настроек оформления. Цвет линии
plt.plot(X,Y, 'k-')
plt.plot(X,Y_noise, 'r+')
```

```python
Y2 = X**2
```

```python
# Короткая форма записи для настроек оформления. Цвет линии
plt.plot(X,Y, 'k-')
plt.plot(X,Y_noise, 'r+')
plt.plot(X[::9],Y2[::9], 'go')
```

```python
# Полная форма записи для настроек оформления. Цвет, тип и толщина линии
plt.plot(X,Y_noise, color='red', linestyle='-',
        linewidth=3, markersize=3)
plt.plot(X,Y, color='black',  linestyle='-',
        linewidth=6, markersize=3)
plt.plot(X[::9],Y2[::9], color='green', marker='o', linestyle='dashed',
        linewidth=2, markersize=12)
```
