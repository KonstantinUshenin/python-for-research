```python
# Импортируем модули
# numpy - удобная и быстрая работа с массивами
# matplotlib - библиотека для построения графиков
import numpy as np
import matplotlib.pylab as plt
```

```python
# Вывод версий библиотек
np.__version__, plt.__version__
```

```python
# Загрузка картинки и вывод ее в виде текста
img_origin = plt.imread('./rabbit.jpg')
img_origin
```

```python
# Вывод изображения в блокноте
plt.imshow(img_origin)
```

```python
# Свойства изображения как массива
# (высота, ширина, количество цветовых каналов)
# RGB - Red, Green, Blue
img_origin.shape
```

```python
# Синтаксис срезов массива (англ. array slice)
plt.imshow(img_origin[:,:,:])
```

```python
# Вырезание изображения (англ. image crop)
plt.imshow(img_origin[0:600,:,:])
```

```python
# Вырезание изображения (англ. image crop)
plt.imshow(img_origin[0:600,190:550,:])
```

```python
# Изменение размеров (англ. resize)
# Работает за счет взятия N-ного пикселя
# ВНИМАНИИЕ! Для практических приложений лучше использовать scipy.ndimage.zoom
plt.imshow(img_origin[::2,::2,:])
```

```python
# изменение размеров (англ. resize)
# работает за счет взятия N-ного пикселя
# ВНИМАНИИЕ! Для практических приложений лучше использовать scipy.ndimage.zoom
plt.imshow(img_origin[::20,::20,:])
```

```python
# зеркальное отображение (англ. mirror flip)
plt.imshow(img_origin[::-1,:,:])
```

```python
# зеркальное отображение (англ. mirror)
plt.imshow(img_origin[:,::-1,:])
```

```python
# Комбинация обрезки изображения, изменения размера и зеркальных отображений
plt.imshow(img_origin[0:600:15,190:550:15,:][::-1,::-1,:])
```
