```python
# Импортируем модули
# numpy - удобная и быстрая работа с массивами
# matplotlib - библиотека для построения графиков
import numpy as np
import matplotlib.pylab as plt
```

```python
# Вывод версий библиотек
np.__version__, plt.__version__
```

```python
# Зададим графики функций для визуализации
X = np.linspace(-10,10,100)
Y1 = X
Y2 = X**2
Y3 = X**3

Y3_noise = Y3 + np.random.normal(0,100,100)
```

```python
# Оформление легенды графика
plt.plot(X,Y1, label='x')
plt.plot(X,Y2, label='x^2')
plt.plot(X,Y3, label='x^3')
plt.legend()
```

```python
# Оформление легенды графика с подписями в виде latex-нотации
# ВНИМАНИЕ! ЭТО МОЖЕТ НЕ РАБОТАТЬ! 
# Так как не у всех на компьютерах есть интерпретаторы latex.
# 
# Если выдает ошибку, то
# ищите ошибку в google или yandex, а
# затем следуйте инструкциям
#
plt.plot(X,Y1, label='$x$')
plt.plot(X,Y2, label='$x^2$')
plt.plot(X,Y3, label='$x^3$')
plt.legend()
```

```python
# Заголовок графика и подписи осей
plt.title('Sample')
plt.xlabel('X-coord')
plt.ylabel('Y-coord')
plt.plot(X,Y1, label='$x$')
plt.plot(X,Y2, label='$x^2$')
plt.plot(X,Y3, label='$x^3$')
plt.legend()
```

```python
# Русские буквы
plt.title('Пример')
plt.xlabel('Ось абсцисс')
plt.ylabel('Ось ординат')
plt.plot(X,Y1, label='$x$')
plt.plot(X,Y2, label='$x^2$')
plt.plot(X,Y3, label='$x^3$')
plt.legend()
```

```python
# Сохранение графиков в файл
plt.title('Пример')
plt.xlabel('Ось абсцисс')
plt.ylabel('Ось ординат')
plt.plot(X,Y1, label='$x$')
plt.plot(X,Y2, label='$x^2$')
plt.plot(X,Y3, label='$x^3$')
plt.legend()
plt.savefig('sample.jpg')
plt.savefig('sample.png')
plt.savefig('sample.eps')
plt.savefig('sample.pdf')
```

```python
# Сохранение графиков в файл
csfont = {'fontsize':18}

plt.title('Пример', **csfont)
plt.xlabel('Ось абсцисс', **csfont)
plt.ylabel('Ось ординат', **csfont)
plt.plot(X,Y1, label='$x$')
plt.plot(X,Y2, label='$x^2$')
plt.plot(X,Y3, label='$x^3$')
plt.legend(**csfont)
plt.savefig('sample.jpg')
plt.savefig('sample.png')
plt.savefig('sample.eps')
plt.savefig('sample.pdf')
```

```python
# Отдельный размер шрифтов для подписей и для меток
ticksfont = {'fontsize':14}
csfont = {'fontsize':18}

plt.title('Пример', **csfont)
plt.xlabel('Ось абсцисс', **csfont)
plt.ylabel('Ось ординат', **csfont)
plt.plot(X,Y1, label='$x$')
plt.plot(X,Y2, label='$x^2$')
plt.plot(X,Y3, label='$x^3$')
plt.legend(**csfont)
plt.xticks(**ticksfont)
plt.yticks(**ticksfont)
plt.savefig('sample.jpg')
plt.savefig('sample.png')
plt.savefig('sample.eps')
plt.savefig('sample.pdf')
```

```python
# Более жирные линии
# Идеальный вариант, чтобы сделать графики для презентации
ticksfont = {'fontsize':14}
csfont = {'fontsize':18}

plt.title('Пример', **csfont)
plt.xlabel('Ось абсцисс', **csfont)
plt.ylabel('Ось ординат', **csfont)
plt.plot(X,Y1, label='$x$', linewidth='5')
plt.plot(X,Y2, label='$x^2$', linewidth='5')
plt.plot(X,Y3, label='$x^3$', linewidth='5')
plt.legend(**csfont)
plt.xticks(**ticksfont)
plt.yticks(**ticksfont)
plt.savefig('sample.jpg')
plt.savefig('sample.png')
plt.savefig('sample.eps')
plt.savefig('sample.pdf')
```
