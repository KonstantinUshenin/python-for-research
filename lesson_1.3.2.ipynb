#### Занятие 1.3.2 | Каналы изображения. Изображения в градации серого

Содержание: 
- Просмотр каждого цветового канала по отдельности
- Изображение в градации серого

```python
# Импортируем модули
# numpy - удобная и быстрая работа с массивами
# matplotlib - библиотека для построения графиков
import numpy as np
import matplotlib.pylab as plt
```

```python
# Вывод версий библиотек
np.__version__, plt.__version__
```

```python
# Загрузка изображения и ее вывод
img_origin = plt.imread('./rabbit.jpg')
plt.imshow(img_origin)
```

```python
# Первый канал изображения (красный)
img_red = np.zeros_like(img_origin)
img_red[:,:,0] = img_origin[:,:,0]
plt.imshow(img_red)
```

```python
# Второй канал изображения (зеленый)
img_red = np.zeros_like(img_origin)
img_red[:,:,1] = img_origin[:,:,1]
plt.imshow(img_red)
```

```python
# Третий канал изображения (синий)
img_red = np.zeros_like(img_origin)
img_red[:,:,2] = img_origin[:,:,2]
plt.imshow(img_red)
```

```python
# Градация серого. По умолчанию применяется цветовая схема viridis
img_gray1 = 0.3333*img_origin[:,:,0] + 0.3333*img_origin[:,:,1] + 0.3333*img_origin[:,:,2]
plt.imshow(img_gray1)
plt.colorbar()
```

```python
# Градация серого с корректным отображением
img_gray1 = 0.3333*img_origin[:,:,0] + 0.3333*img_origin[:,:,1] + 0.3333*img_origin[:,:,2]
plt.imshow(img_gray1, cmap='gray')
plt.colorbar()
```

```python
# Изображение в градации серого с цветовой схемой jet
img_gray1 = 0.3333*img_origin[:,:,0] + 0.3333*img_origin[:,:,1] + 0.3333*img_origin[:,:,2]
plt.imshow(img_gray1, cmap='jet')
plt.colorbar()
```

```python
# Тестирование разных способов задания градации серого
img_gray1 = 0.3333*img_origin[:,:,0] + 0.3333*img_origin[:,:,1] + 0.3333*img_origin[:,:,2]
plt.imshow(img_gray1, cmap='gray')
plt.show()
img_gray2 = 0.299*img_origin[:,:,0] + 0.587*img_origin[:,:,1] + 0.114*img_origin[:,:,2]
plt.imshow(img_gray2, cmap='gray')
plt.show()
img_gray3 = 0.2627*img_origin[:,:,0] + 0.6780*img_origin[:,:,1] + 0.0593*img_origin[:,:,2]
plt.imshow(img_gray2, cmap='gray')
plt.show()
```
